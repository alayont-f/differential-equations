In this activity, we will learn about a graphical representation of two-dimensional systems (of differential equations). In other words, we will focus on systems where we have two variables $x, y$, each depending on $t$. This graphical representation is similar to the phase line from the one-dimensional case.

For a two-dimensional autonomous system (i.e. where $t$ does not appear explicitly in the differential equations), we use a 2-D picture to represent the behaviour of the system. Suppose our dependent variables are $x$ and $y$. Similar to the phase line corresponding to the $y$-values, the **phase plane** consists of the $x$ and $y$-values (and hence the name plane). Different than the phase line, we do not place arrows in the phase plane to indicate general direction of the movement (we will place arrows later below in a different graphical representation). Also different than the phase line, it makes sense to plot trajectories (also called solution curves) of variable values representing how $x$ and $y$ change with time on the phase plane. By placing a few trajectories on the phase plane, we can see the general behaviour of the system for various initial $x$ and $y$ values. This whole picture will be called the **phase portrait**. 

Python has a pre-made command that does all of this for us as shown in the code below.

```
from pylab import *
#set viewing window
xbegin = -5
xend = 5
xby = 0.5
ybegin = -5
yend = 5
yby = 0.5

x, y = meshgrid(arange(xbegin, xend, xby), arange(ybegin, yend, yby))
dx = -3*x + 10*y           #change these two lines for a different system
dy = -1*x +3*y

streamplot(x, y, dx, dy)
grid(); show()
```

Sometimes though, we would like to see how in general solutions would behave in an area using a slope field like tool. For systems of differential equations, the slope field becomes a **vector field**. We can visualize this vector field for an autonomous system of differential equations in two variables by using a 2-D graph. 

Let us first define what a vector field is. In the case we are interested, we will obtain a two-dimensional vector field on a subset of ${\bf R}^2$ (or all of ${\bf R}^2$). This vector field will assign to each $(x,y)$ point in this subset a two-dimensional vector. Given a system of differential equations in $x$ and $y$, at an $(x,y)$ point in the $xy$-plane, we assign the vector $\begin{bmatrix} \frac{dx}{dt} \\ \frac{dy}{dt} \end{bmatrix}$ (whenever it is defined). This vector represents how fast each of $x$ and $y$ will changing (with respect to time) if we start at those specific $x$ and $y$ values. In other words, this vector is similar to the slope values we calculated when we used the Euler's method for single variable equations.

If we want to numerically estimate the solution curves, we need the lengths of the vectors in the vector field. However, in order to guess the shape of the solution curves qualitatively, we only need the direction of these vectors. So, the picture we obtain by scaling the vectors in the vector field so that they all have the same length is called the **direction field**. This also helps us see the direction of the vectors with very small magnitudes.

Below is a code to generate a direction field.

(Credit: This portion below is a slightly modified version of an activity by Dr. Brian Drake.)

```
import numpy as np
import matplotlib.pyplot as plt

#set viewing window
xstart = -10
xend = 10
ystart = -10
yend = 10
xby = 1  #change this to get arrows closer in the x-direction
yby = 1  #change this to get arrows closer in the y-direction

x,y = np.meshgrid(np.arange(xstart,xend,xby),
                    np.arange(ystart,yend,yby))

#place the system here. dx should be equal to dx/dt of your system, dy=dy/dt.
dx = 0*x+1*y        #this is system dx/dt=y and dy/dt=-x
dy = -1*x-0*y         

#correct directions to match scale of plot
dxb = dx/(xend-xstart)
dyb = dy/(yend-ystart)

#normalize lengths of arrows   
#(uncomment next two lines for a direction field, which uses unit length arrows)
#length = np.sqrt(dxb**2+dyb**2)
#u, v = dxb/length, dyb/length

#don't normalize lengths of arrows  
#(uncomment the beginning of next line for vector field, which doesn't 
#normalize vectors)
u, v = dxb, dyb                     

#create the plot
width=10    #change plot size if the length does not equal width here
height=10
fig, ax = plt.subplots(figsize=(width,height))

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('System of Two Variables',fontsize=20)
ax.set_xlim(xstart,xend)
ax.set_ylim(ystart,yend)
ax.grid(True, linewidth=0.2)

ax.quiver(x,y,u*width,v*height, color='g')

plt.show()
```
